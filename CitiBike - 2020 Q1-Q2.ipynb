```python
# Dependencies & Setup
import pandas as pd
```

```python
# Files to Load
# Files from Citi Bike trip history data, Source: https://s3.amazonaws.com/tripdata/index.html
jan_2020 = "data/202001-citibike-tripdata.csv"
feb_2020 = "data/202002-citibike-tripdata.csv"
mar_2020 = "data/202003-citibike-tripdata.csv"
apr_2020 = "data/202004-citibike-tripdata.csv"
may_2020 = "data/202005-citibike-tripdata.csv"
jun_2020 = "data/202006-citibike-tripdata.csv"

# Read Quarter 1 and Quarter 2 Citi Bike Data Files & Store into Pandas DataFrames
df_jan_2020 = pd.read_csv(jan_2020, encoding="ISO-8859-1")
df_feb_2020 = pd.read_csv(feb_2020, encoding="ISO-8859-1")
df_mar_2020 = pd.read_csv(mar_2020, encoding="ISO-8859-1")
df_apr_2020 = pd.read_csv(apr_2020, encoding="ISO-8859-1")
df_may_2020 = pd.read_csv(may_2020, encoding="ISO-8859-1")
df_jun_2020 = pd.read_csv(jun_2020, encoding="ISO-8859-1")
```

```python
# df_jan_2020 
# df_feb_2020 
# df_mar_2020
# df_apr_2020 
# df_may_2020 
# df_jun_2020 
```

```python
# Combining all csv files, since no unique join column id
combined_citibike_df = df_jan_2020.append(df_feb_2020, sort=False)
combined_citibike_df = combined_citibike_df.append(df_mar_2020, sort=False)
combined_citibike_df = combined_citibike_df.append(df_apr_2020, sort=False)
combined_citibike_df = combined_citibike_df.append(df_may_2020, sort=False)
combined_citibike_df = combined_citibike_df.append(df_jun_2020, sort=False)

combined_citibike_df
```

```python
# Replacing Numbers in Gender for Male, Female, Unknown

combined_citibike_df['gender'] = combined_citibike_df ['gender'].replace([0,1,2],['Unknown','Male','Female'])
combined_citibike_df
```

```python
# Checking to see count of datapoints
combined_citibike_df.count()
```

```python
# Dropping NaN
combined_citibike_df = combined_citibike_df.dropna(how='any')
combined_citibike_df.count()
```
